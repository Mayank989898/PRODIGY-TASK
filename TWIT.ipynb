```python
plt.figure(figsize=(10,10))
df['Borderlands'].value_counts().plot(kind='barh',color='b')
plt.xlabel("Count")
plt.show()
```

```python
sns.countplot(x = 'Borderlands',data=df,palette='viridis')
plt.show()
```

```python
# Calculate the counts for each Borderlands
Borderlands_counts = df['Borderlands'].value_counts()

# Create the pie chart
plt.figure(figsize=(10, 10))
plt.pie(Borderlands_counts, labels=Borderlands_counts.index, autopct="%1.1f%%", startangle=140, colors=['skyblue', 'orange', 'green', 'red', 'purple'])

plt.title('Borderlands Distribution')

# Show the plot
plt.show()
```

```python
df['msg_len'] = df['im getting on borderlands and i will murder you all ,'].apply(len)
```

```python
df
```

```python
sns.histplot(df['msg_len'], bins=25,kde=True)
plt.title('Message Length Distribution in Training Data')
plt.ylabel('Frequency')
plt.xlabel('Message Length')
plt.show()
```

```python
sns.boxplot(data=df, x=df['Positive'], y='msg_len', palette='viridis', order=['Positive', 'Negative', 'Neutral', 'Irrelevant'])
plt.title('Message Length Distribution by Sentiment in Training Data')
plt.ylabel('Message Length')
plt.xlabel('Positive')
plt.ylim(0,300)
plt.show()
```

```python

crosstab = pd.crosstab(index=df['Borderlands'], columns=df['Positive'])

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(crosstab, cmap='coolwarm', annot=True, fmt='d', linewidths=.5)

# Add labels and title
plt.title('Heatmap of Borderlands vs Positive')
plt.xlabel('Positive')
plt.ylabel('Borderlands')

# Show the plot
plt.show()
```
